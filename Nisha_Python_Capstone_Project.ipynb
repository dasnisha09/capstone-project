## Task 2
#### Replacing missing values by running average

The cost column in dataframe “Project” has some missing values. Your task is to
compute these missing values. Replace the missing values by running average. You
should use “For” loop for this task.

> Note: I have tried using "for" loop but I somehow could not use it efficiently so, finally went ahead by manually replacing the missing values.

```python
#Importing the project file as dataframe

project_df = pd.read_csv('project_data.csv')
project_df
```

```python
#Creating a list of running averages using 'Cost' column in the project dataframe
running_averages = []
running_sum = 0
count = 0

for cost in project_df['Cost']:
    if not pd.isna(cost):
        running_sum += cost
        count += 1
    if count > 0:
        running_avg = running_sum / count
        running_averages.append(running_avg)
    else:
        running_averages.append(None)
```

```python
#inserting the Running Average Series as a column in the dataframe beside the 'Cost' column.
project_df['Running_Average'] = running_averages
```

```python
#Using fillna() backward fill method to fill NaN values with the running averages.
project_df['Cost'].fillna(method='bfill', inplace=True)
```

```python
#Since we no longer need the running averages, I chose to drop that column from the dataframe.
project_df.to_csv('project_with_running_avg.csv', index=False)
running_averages
```

Add a new bonus column in Final dataframe. Give 5% bonus concerning project cost
only for employees who have finished the projects.

```python
#Using 'for' loop to creating a list of Bonus cost
#Later, inserted the list into the final dataframe

final_df = pd.read_csv('project_data.csv')
```

```python
# Initialize an empty list to store bonus values
bonus_list = []
# Iterate through each row in the 'final' DataFrame
for index, row in final_df.iterrows():
    if row['Status'] == 'Finished':
        bonus = row['Cost'] * 0.05
    else:
        bonus = 0
    bonus_list.append(bonus)

final_df['Bonus-5%'] = bonus_list
bonus_list
```

## Task 9

Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost
> Adding the cost of all projects for each Employee and saving it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost

```python
#Firstly, I grouped the 'final' dataframe using ID and First Name columns
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv("employee_data.csv")
merged_df = pd.merge(employees_df, project_df, on='ID')
merged_df
```

```python
#Used sum() function to aggregate the group to find out the total cost of each employee's project
TotalProjCost = merged_df.groupby(['ID', 'Name'])['Cost'].sum().reset_index()
TotalProjCost.rename(columns={'Cost': 'Total cost'}, inplace=True)
```
